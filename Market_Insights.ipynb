```python
# Understand the retention of customers on month on month basis

import pandas as pd

# Sample DataFrame (make sure you already have the 'online_sales' dataframe)
# online_sales = pd.read_csv('your_data.csv')  # load your actual data

# Convert Transaction_Date to datetime
online_sales['Transaction_Date'] = pd.to_datetime(online_sales['Transaction_Date'])

# Get the first purchase date for each customer (acquisition month)
customer_acquisition = online_sales.groupby('CustomerID')['Transaction_Date'].min().reset_index()
customer_acquisition['Acquisition_Month'] = customer_acquisition['Transaction_Date'].dt.to_period('M')

# Add Acquisition_Month to the original sales data
online_sales['Transaction_Month'] = online_sales['Transaction_Date'].dt.to_period('M')

# Identify the customers who made purchases in each month
customers_per_month = online_sales.groupby('Transaction_Month')['CustomerID'].nunique().reset_index(name='Customer_Count')

# For each customer, check if they made a purchase in months after their acquisition
customer_retention = customer_acquisition.copy()

# Track retention: for each month after acquisition, check if the customer made a purchase
retention_data = []

for month in customer_retention['Acquisition_Month'].unique():
    # Filter customers acquired in this month
    acquired_customers = customer_retention[customer_retention['Acquisition_Month'] == month]

    for m in online_sales['Transaction_Month'].unique():
        # Count how many acquired customers made a purchase in month 'm'
        retained_customers = online_sales[(online_sales['CustomerID'].isin(acquired_customers['CustomerID'])) &
                                          (online_sales['Transaction_Month'] == m)]
        retention_count = retained_customers['CustomerID'].nunique()

        retention_data.append({
            'Acquisition_Month': month,
            'Retention_Month': m,
            'Retained_Customers': retention_count,
            'Retention_Rate': retention_count / acquired_customers['CustomerID'].nunique()
        })

# Create a DataFrame for retention data
retention_df = pd.DataFrame(retention_data)

# Calculate the retention rate: retained customers divided by total customers acquired in the acquisition month
#retention_df['Retention_Rate'] = retention_df['Retained_Customers'] / retention_df.groupby('Acquisition_Month')['Retained_Customers'].transform('first')

```

```python
'''Performing Customer Segmentation
 Heuristic (Value based, RFM) – Divide the customers into Premium, Gold, Silver,
Standard customers and define strategy on the same. '''


import pandas as pd
import numpy as np
from datetime import datetime

# Sample data
# sales_data = pd.read_csv('your_sales_data.csv')  # Replace with your actual data

# Convert 'Transaction_Date' to datetime
sales_with_tax_coupon_customer['Transaction_Date'] = pd.to_datetime(sales_with_tax_coupon_customer['Transaction_Date'])

# Get the latest transaction date to calculate Recency
latest_date = sales_with_tax_coupon_customer['Transaction_Date'].max()
```

```python
sales_with_tax_coupon_customer.head(1)
```

```python
# Calculate Recency: Days since last purchase
rfm = sales_with_tax_coupon_customer.groupby('CustomerID').agg({
    'Transaction_Date': lambda x: (latest_date - x.max()).days,  # Recency
    'Transaction_ID': 'count',  # Frequency
    'Invoice_Value': 'sum'  # Monetary
}).reset_index()
```

```python
# Rename columns
rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']

# Normalize the RFM variables (scale them between 1 to 5)
rfm['Recency_Score'] = pd.qcut(rfm['Recency'], 5, labels=False) + 1
rfm['Frequency_Score'] = pd.qcut(rfm['Frequency'], 5, labels=False) + 1
rfm['Monetary_Score'] = pd.qcut(rfm['Monetary'], 5, labels=False) + 1
```

```python
# Calculate the total score

rfm['RFM_Score'] = rfm['Recency_Score'] + rfm['Frequency_Score'] + rfm['Monetary_Score']
```

```python
# Define segments based on the RFM score

def assign_segment(rfm_score):
    if rfm_score >= 13:
        return 'Premium'
    elif rfm_score >= 10:
        return 'Gold'
    elif rfm_score >= 7:
        return 'Silver'
    else:
        return 'Standard'

rfm['Segment'] = rfm['RFM_Score'].apply(assign_segment)

# Display the segmented customers
print(rfm[['CustomerID', 'Recency_Score', 'Frequency_Score', 'Monetary_Score', 'RFM_Score', 'Segment']])
```

```python
# Scientific (Using K-Means) & Understand the profiles. Define strategy for each
# segment.

import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'rfm' contains the 'Recency', 'Frequency', and 'Monetary' values for each customer
# Normalize the data
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm[['Recency', 'Frequency', 'Monetary']])

# Determine the optimal number of clusters (using the Elbow Method)
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(rfm_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow curve
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()

# Based on the Elbow method, assume the optimal number of clusters is 4
kmeans = KMeans(n_clusters=4, random_state=42)
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)

# Analyzing the clusters: Get the cluster centroids
cluster_centroids = pd.DataFrame(scaler.inverse_transform(kmeans.cluster_centers_), columns=['Recency', 'Frequency', 'Monetary'])
print("Cluster Centroids:\n", cluster_centroids)

# Visualizing the clusters
sns.scatterplot(x='Recency', y='Frequency', hue='Cluster', data=rfm, palette='viridis', s=100)
plt.title('Customer Segments (Recency vs Frequency)')
plt.show()

# Define strategies for each segment
segment_strategies = {
    0: "High Frequency, Low Recency (Loyal customers) - Engage with loyalty programs.",
    1: "Low Frequency, Low Recency (Dormant customers) - Re-engagement campaigns.",
    2: "High Frequency, High Monetary (High-value customers) - VIP programs and offers.",
    3: "Low Frequency, High Monetary (Big spenders, infrequent) - Targeted promotions to increase frequency."
}

# Add the segment strategies to the dataframe for each customer
rfm['Segment_Strategy'] = rfm['Cluster'].map(segment_strategies)

# Show the final dataframe with customer segments and strategies
print(rfm.head())
```

```python
import pandas as pd
from itertools import combinations
from collections import Counter

# Step 1: Prepare the data
# Create a DataFrame where each transaction contains the list of purchased products
transaction_products = sales_with_tax_coupon_customer.groupby('Transaction_ID')['Product_SKU'].apply(list)

# Step 2: Generate product combinations
# Define a function to generate combinations of a specific size
def generate_combinations(data, size):
    combo_counts = Counter()
    for products in data:
        if len(products) >= size:
            combo_counts.update(combinations(sorted(products), size))
    return combo_counts

# Step 3: Find the most frequent combinations for sizes 1 to 5
top_combos = {}
for size in range(1, 6):  # Sizes from 1 to 5
    combos = generate_combinations(transaction_products, size)
    top_combos[size] = combos.most_common(1)

# Step 4: Display results
for size, combo in top_combos.items():
    print(f"Top {size}-Product Combo: {combo}")
```

```python
'''Predicting Next Purchase Day(How soon each customer can visit the store (0-30 days, 30-60
days, 60-90 days, 90+ days)
 For this, we need create dependent variable at customer level (average days per one
transaction for only repeat customers and divide into groups 0-30 days, 30-60 days,
60-90 days and 90+ days) then build classification model to predict next purchase of
given customer.'''

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import warnings

warnings.filterwarnings("ignore")

# Step 1: Prepare Data
# Assuming `sales_with_tax_coupon_customer` contains `CustomerID`, `Transaction_Date`
sales_with_tax_coupon_customer['Transaction_Date'] = pd.to_datetime(sales_with_tax_coupon_customer['Transaction_Date'])

# Calculate days between transactions for each customer
sales_with_tax_coupon_customer['Days_Between'] = sales_with_tax_coupon_customer.groupby('CustomerID')['Transaction_Date'].diff().dt.days

# Filter only repeat customers
repeat_customers = sales_with_tax_coupon_customer.dropna(subset=['Days_Between'])

# Calculate average days per transaction for repeat customers
customer_avg_days = repeat_customers.groupby('CustomerID')['Days_Between'].mean().reset_index()
customer_avg_days.rename(columns={'Days_Between': 'Avg_Days_Between'}, inplace=True)

# Step 2: Create Dependent Variable
def classify_days(days):
    if days <= 30:
        return 0  # 0-30 days
    elif days <= 60:
        return 1  # 30-60 days
    elif days <= 90:
        return 2  # 60-90 days
    else:
        return 3  # 90+ days

customer_avg_days['Next_Purchase_Group'] = customer_avg_days['Avg_Days_Between'].apply(classify_days)

# Merge features from sales data (e.g., total revenue, frequency)
customer_features = sales_with_tax_coupon_customer.groupby('CustomerID').agg({
    'Invoice_Value': 'sum',  # Total Revenue
    'Transaction_Date': 'count'  # Frequency
}).reset_index()
customer_features.rename(columns={'Invoice_Value': 'Total_Revenue', 'Transaction_Date': 'Frequency'}, inplace=True)

# Combine data
data = pd.merge(customer_avg_days, customer_features, on='CustomerID')

# Step 3: Build Classification Model
X = data[['Total_Revenue', 'Frequency']]
y = data['Next_Purchase_Group']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Random Forest Classifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Step 4: Evaluate Model
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Step 5: Display Feature Importance
feature_importances = pd.DataFrame(model.feature_importances_, index=X.columns, columns=['Importance']).sort_values(by='Importance', ascending=False)
print("\nFeature Importances:\n", feature_importances)
```

```python
'''Perform cohort analysis by defining below cohorts
 Customers who started in each month and understand their behaviour
 Which Month cohort has maximum retention?'''

import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'sales_with_tax_coupon_customer' is your DataFrame containing sales data
# Step 1: Extract the first purchase month for each customer
sales_with_tax_coupon_customer['Transaction_Date'] = pd.to_datetime(sales_with_tax_coupon_customer['Transaction_Date'])

# Get the first purchase date for each customer
first_purchase = sales_with_tax_coupon_customer.groupby('CustomerID')['Transaction_Date'].min().reset_index()
first_purchase['First_Purchase_Month'] = first_purchase['Transaction_Date'].dt.to_period('M')

# Step 2: Merge with the original dataset to assign the first purchase month to each transaction
sales_with_tax_coupon_customer = pd.merge(sales_with_tax_coupon_customer, first_purchase[['CustomerID', 'First_Purchase_Month']], on='CustomerID')

# Step 3: Create a "Cohort Period" which represents how many months after the first purchase the transaction occurred
sales_with_tax_coupon_customer['Cohort_Period'] = (sales_with_tax_coupon_customer['Transaction_Date'].dt.to_period('M') - sales_with_tax_coupon_customer['First_Purchase_Month']).apply(lambda x: x.n)

# Step 4: Calculate retention by cohort (percentage of customers making a purchase in each cohort period)
cohort_data = sales_with_tax_coupon_customer.groupby(['First_Purchase_Month', 'Cohort_Period'])['CustomerID'].nunique().reset_index()

# Total customers in each cohort
cohort_sizes = cohort_data[cohort_data['Cohort_Period'] == 0][['First_Purchase_Month', 'CustomerID']].rename(columns={'CustomerID': 'Cohort_Size'})

# Merge cohort sizes with the main cohort data
cohort_data = pd.merge(cohort_data, cohort_sizes, on='First_Purchase_Month')

# Step 5: Calculate retention rate (number of customers in a cohort for each period / cohort size)
cohort_data['Retention_Rate'] = cohort_data['CustomerID'] / cohort_data['Cohort_Size']

# Step 6: Pivot the table to create a cohort analysis table
cohort_pivot = cohort_data.pivot_table(index='First_Purchase_Month', columns='Cohort_Period', values='Retention_Rate')
```

```python

# Step 7: Visualize the cohort analysis with retention rates in each box
plt.figure(figsize=(12, 8))
plt.title('Cohort Analysis - Retention Rates')
plt.xlabel('Cohort Period (Months)')
plt.ylabel('Cohort Start Month')
plt.imshow(cohort_pivot, cmap='Blues', aspect='auto')
plt.colorbar()
plt.xticks(ticks=range(len(cohort_pivot.columns)), labels=cohort_pivot.columns)
plt.yticks(ticks=range(len(cohort_pivot.index)), labels=cohort_pivot.index)

# Annotate each box with the retention rate
for i in range(len(cohort_pivot.index)):
    for j in range(len(cohort_pivot.columns)):
        retention_rate = cohort_pivot.iloc[i, j]
        if pd.notna(retention_rate):  # Only annotate non-NaN values
            plt.text(j, i, f'{retention_rate:.2f}', ha='center', va='center', color='black')

plt.show()

# Step 8: Find which cohort has the maximum retention
max_retention_cohort = cohort_pivot.max(axis=1).idxmax()
print(f"The cohort with the highest retention is from {max_retention_cohort}.")
```
